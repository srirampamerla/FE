# Target Guided Ordinal Encoding

1.Ordering the labels according to the target

2.Replace the labels by the joint probability of being 1 or 0

```python
import pandas as pd
df=pd.read_csv('titanic_train.csv', usecols=['Cabin','Survived'])
df.head()
```

```python
# df['Cabin'].fillna('Missing',inplace=True)
import numpy as np
df['Cabin']=np.where(df['Cabin'].isnull(),'Missing',df['Cabin'])
```

```python
df.head()
```

```python
df['Cabin'].dtypes
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]
```

```python
df['Cabin'].dtypes
```

```python
df.head()
```

```python
df.Cabin.unique()
```

```python
df.groupby(['Cabin'])['Survived'].mean()
```

```python
df.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
df.groupby(['Cabin'])['Survived'].mean().sort_values().values
```

```python
ordinal_labels=df.groupby(['Cabin'])['Survived'].mean().sort_values().index
ordinal_labels
```

```python
import pandas as pd
import numpy as np

def one_hot_encoding(df, variable, top_x_label):
    # Ensure top_x_labels is a list of top categories for the variable
    top_x_labels = df[variable].value_counts().sort_values(ascending=False).head(top_x_label).index
    top_x_labels = list(top_x_labels)
    
    for category in top_x_labels:
        df[variable+'_'+category] = np.where(df[variable] == category, 1, 0)
    
    return df

# Read the data
df = pd.read_csv('mercedes.csv', usecols=['X1', 'X2', 'X3', 'X4', 'X5', 'X6'])

# Specify the number of top labels to encode
top_x_labels = 10

# Apply one-hot encoding
df = one_hot_encoding(df, 'X2', top_x_labels)
df.head()
```

```python
pd.get_dummies(df)
```

```python
import pandas as pd
import numpy as np

def one_hot_encoding(df, variable, top_x_labels):
    # Ensure top_x_labels is a list of top categories for the variable
    top_x_labels = df[variable].value_counts().sort_values(ascending=False).head(top_x_labels).index
    top_x_labels = list(top_x_labels)
    
    for category in top_x_labels:
        df[category] = np.where(df[variable] == category, 1, 0)
    

    
    # Print the output
    print(df[[variable] + top_x_labels].head(20))
    
    return df

# Read the data
df = pd.read_csv('mercedes.csv', usecols=['X1', 'X2', 'X3', 'X4', 'X5', 'X6'])

# Specify the number of top labels to encode
top_x_labels = 10

# Apply one-hot encoding
df = one_hot_encoding(df, 'X2', top_x_labels)
```

```python
lst_10=df.X1.value_counts().sort_values(ascending=False).head(10).index
lst_10=list(lst_10)
lst_10
import numpy as np
for categories in lst_10:
    df[categories]=np.where(df['X2']==categories,1,0)
    # replacing however the category is matching the category i the column hat will be one remaining zero
df[['X2'] + lst_10].head(20)   
```

```python
# label Encoding
import pandas as pd
from sklearn.preprocessing import LabelEncoder

def label_encoding(df, variable):
    # Initialize the label encoder
    le = LabelEncoder()
    
    # Fit and transform the variable
    df[variable + '_encoded'] = le.fit_transform(df[variable])
    
    # Print the output
    print(df[['X1', variable, variable + '_encoded']].head(20))
    
    return df

# Read the data
df = pd.read_csv('mercedes.csv', usecols=['X1', 'X2', 'X3', 'X4', 'X5', 'X6'])

# Apply label encoding
df = label_encoding(df, 'X2')
```
